```python
import csv

def append_to_csv(data):
    # Open the CSV file in append mode
    with open('data.csv', 'a', newline='', encoding='utf-8') as csvfile:
        # Create a CSV writer
        writer = csv.DictWriter(csvfile, fieldnames=['title', 'article', 'fake', 'category'])

        # Write the header row only if the file is empty
        if csvfile.tell() == 0:
            writer.writeheader()

        # Write each row of data
        for row in data:
            writer.writerow(row)

append_to_csv(data)
```
