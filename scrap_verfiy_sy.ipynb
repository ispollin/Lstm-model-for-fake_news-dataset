```python
from bs4 import BeautifulSoup
import requests
import pandas as pd

# Base URL of the website to scrape
base_url = 'https://verify-sy.com/all/18/%D9%83%D8%B0%D8%A8?page={}'

# Function to scrape paragraphs before "news-details-sections" class within "wekek" class
def scrape_paragraphs(url, title):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Parse the HTML content
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # Find the "wekek" class
    wekek_div = soup.find(class_='wekek')
    
    paragraphs_list = []
    
    if wekek_div:
        # Find the "news-details-sections" class within "wekek" class
        news_details_sections_div = wekek_div.find(class_='news-details-sections')
        
        # Check if "news-details-sections" class is found
        if news_details_sections_div:
            # Find all preceding siblings paragraphs
            paragraphs = news_details_sections_div.find_previous_siblings('p')
            
            # Extract and print or store the paragraphs
            for paragraph in paragraphs:
                paragraphs_list.append(paragraph.text.strip())

        else:
            print("Could not find 'news-details-sections' class within 'wekek' class on the page.")
    else:
        print("Could not find 'wekek' class on the page.")
    
    return paragraphs_list

# Create lists to store titles and paragraphs
titles = []
paragraphs = []

# Iterate through pages from 1 to 21
for i in range(1, 22):
    # Construct the URL for the current page
    url = base_url.format(i)
    
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if request was successful
    if response.status_code == 200:
        # Parse the HTML content
        soup = BeautifulSoup(response.text, 'html.parser')
        
        # Find the specific section containing the articles
        divs = soup.find_all('div', class_='blog_food_bigdiv wow fadeInUp col-lg-4 col-12')
        
        # If the section is found, proceed with extracting article information
        if divs:
            # Iterate over each article within the section
            for div in divs:
                # Extract article title
                title = div.find('h3').text.strip()
                titles.append(title)
                
                # Extract article link
                article_link = div.find('a')['href']
                
                # Scrape paragraphs before "news-details-sections" class within "wekek" class
                paragraphs_list = scrape_paragraphs(article_link, title)
                paragraphs.append(' '.join(paragraphs_list))
        else:
            print('Article section not found on page', i)
    else:
        print("Failed to retrieve page", i, "- Status Code:", response.status_code)

# Create DataFrame
df = pd.DataFrame({'title': titles, 'paragraph': paragraphs})

# Display DataFrame
print(df)
```
